```r
model <- lm(Healthy.life.expectancy.at.birth ~ Social.support, data = whr_data)
```

```r
new_data <- data.frame(Social.support = 0.7213)
```

```r
new_data
```

```r
predict(model, new_data, interval = "confidence")
```

```r
ggplot (aes (x = Social.support, y = Healthy.life.expectancy.at.birth), data = whr_data)  + # these represent the x and y variables, as well as the source of the data
  geom_point(alpha = 0.5)+ # changes how transparent the dots are
  geom_smooth(method = lm, se = TRUE, color = "red", fill = "orange")+ # adds in a line of best fit and the confidence interval for that line
  labs(x = "Social Support Score", y = "Life Expectancy (Years)")+ # changes the axis labels
  aes(color = Healthy.life.expectancy.at.birth)+
  theme_classic()+  # removes a lot of the background default ggplot graphics to make a more simple plot
  theme(legend.position = "none")
```
